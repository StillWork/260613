## 1. 챗봇 vs AI 에이전트 — 무엇이 다른가?

| 비교 항목 | 일반 챗봇 | AI 에이전트 |
|-----------|-----------|-------------|
| **동작 방식** | 질문 → 답변 (1회) | 목표 → 계획 → 도구 사용 → 결과 (반복) |
| **도구 사용** | 없음 (언어만) | 검색·계산·파일·API 등 외부 도구 활용 |
| **상태 기억** | 대화 내용만 | 작업 진행 상황, 중간 결과 저장 가능 |
| **자율성** | 사람이 매번 지시 | 목표만 주면 스스로 단계 계획 |
| **대표 예시** | "오늘 날씨 어때?" | "이번 주 뉴스 분석해서 보고서 작성해줘" |

### 핵심 차이 한 줄 요약
> **챗봇**: "답변을 생성하는 AI"  
> **에이전트**: "목표를 달성하기 위해 스스로 행동하는 AI"

### 실제 예시로 비교

**챗봇에게 물어보면:**
```
사용자: 애플 주가 분석해줘
챗봇: 저는 실시간 데이터에 접근할 수 없습니다. 
      일반적으로 주가 분석은 PER, 매출 성장률 등을 봅니다...
```

**에이전트에게 요청하면:**
```
사용자: 애플 주가 분석해줘
에이전트: [search_news("애플 최근 실적")] 실행 중...
          [get_stock_price("AAPL")] 실행 중...
          [calculate_metrics(per=28.5, growth=0.08)] 실행 중...
          분석 결과: 현재 PER 28.5는 업계 평균보다...
```

## 2. 에이전트 작동 원리 — ReAct 루프

에이전트는 **ReAct(Reason + Act)** 패턴으로 작동합니다.

```
사용자 목표 입력
      ↓
┌─────────────────────────────────┐
│  1. Perceive (인식)              │
│     현재 상황과 목표를 파악       │
│              ↓                  │
│  2. Reason (추론)                │
│     다음에 무엇을 해야 할지 판단  │
│              ↓                  │
│  3. Act (행동)                   │
│     도구 선택 → 실행              │
│              ↓                  │
│  4. Observe (관찰)               │
│     결과 확인 → 다시 1번으로      │
└─────────────────────────────────┘
      ↓ (목표 달성 시)
  최종 답변 반환
```

### 도구(Tool)란 무엇인가?

에이전트가 사용할 수 있는 "기능"을 뜻합니다.

| 도구 종류 | 예시 | 활용 상황 |
|-----------|------|-----------|
| **검색** | Google 뉴스 RSS, 웹 검색 | 최신 정보 필요할 때 |
| **계산** | Python 코드 실행, 수식 계산 | 수치 분석 필요할 때 |
| **파일** | 파일 읽기/쓰기, 스프레드시트 | 데이터 저장·불러오기 |
| **API** | 날씨 API, 환율 API, 사내 시스템 | 외부 서비스 연동 |
| **이메일/메시지** | Gmail, Slack 전송 | 결과 공유 |

> 💡 **FM2 시리즈에서 배우는 도구들**: 날씨 계산기(FM2_02) → 뉴스 RSS(FM2_04) → Wikipedia 검색(FM2_05)

### 실습 A: 에이전트처럼 생각하기

```
당신은 시장 조사 에이전트입니다.
다음 목표를 달성하기 위한 단계별 계획을 세우고,
각 단계에서 어떤 도구가 필요한지 명시해주세요.

목표: "국내 AI 스타트업 TOP 5의 최근 투자 현황을 분석하고 
      요약 보고서를 작성해줘"

형식:
- Step 1: [행동] (필요 도구: XXX)
- Step 2: [행동] (필요 도구: XXX)
...
- 최종 결과물: [설명]
```

> 이 프롬프트로 에이전트가 어떻게 계획을 세우는지 직접 확인할 수 있습니다.

### 실습 B: Tool Use 흉내내기

```
당신은 도구를 사용할 수 있는 에이전트입니다.
아래 도구들이 있다고 가정하고, 질문에 답하세요.

사용 가능한 도구:
- search_news(keyword): 뉴스 검색
- get_weather(city): 날씨 조회  
- calculate(expression): 계산

질문: "서울 날씨가 맑고 기온이 28도라면, 
      야외 행사 수익이 실내 대비 30% 높다고 할 때
      예상 추가 수익은 기본 수익 500만원 기준으로 얼마야?"

각 도구를 호출하는 순서와 결과를 보여주세요.
```

## 5. 수강생 유형별 추천 학습 경로

### 나는 어떤 유형인가?

**A. 코딩 경험이 거의 없는 비개발자**
```
FM2_00(지금) → Claude.ai 실습(Lv1) → FM2_03 GAS 파트만 → 업무 적용
```
- 목표: Python 없이도 AI 자동화 업무 적용
- 핵심 예제: FM2_00, FM2_03의 GAS 섹션

**B. Python 기초는 알지만 AI는 처음**
```
FM2_00(지금) → FM2_01 → FM2_02 → FM2_03 or FM2_04 → FM2_05
```
- 목표: API부터 Tool Use까지 순차 학습
- 핵심 예제: FM2_02 (Tool Use가 가장 중요)

**C. 개발 경험이 있고 빠르게 실무 적용하려는 개발자**
```
FM2_00(개념 확인) → FM2_02 → FM2_05 → MCP(Level 5)
```
- 목표: 에이전트 아키텍처 이해 후 실무 프로젝트
- 핵심 예제: FM2_02, FM2_05

**D. 기획/관리자 — 팀에 AI 도입을 검토 중**
```
FM2_00(지금) → Claude.ai 실습만 → 에이전트_교육전략.md 참고
```
- 목표: AI 에이전트의 가능성과 한계를 이해
- 핵심 자료: 이 노트북 + Level0~6 PPT 슬라이드
